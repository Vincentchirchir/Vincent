## Spatial Joins
With spatial joins you can determine for each city the country in which it is located
You achieve this by combining information of geospatial datasets based on their spatial relationship

```python
#To illustrate the concept of joining the information of two dataframes with pandas, I will take a small subset of cities and countries datasets
cities2=cities[cities['name'].isin(['Bern', 'London', 'Brussels', 'Paris'])].copy()
cities2['iso_a3']=['CHE','BEL', 'GBR', 'FRA']
cities2
```

```python
countries2=countries[['iso_a3', 'name', 'continent']]
countries2.head()
```

```python
cities2.merge(countries2, on='iso_a3')
```
